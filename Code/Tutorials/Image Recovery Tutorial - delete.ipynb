The documentation *documentation link* contains more information about parameters.

All parameters have a *default* value to make it easier for the user. 

```python
default_parameters = Im_Rec_Arguments()   # Call default parameters
default_parameters.plot_mode = True       
params = default_parameters.__dict__
params["density"] = 10000
params["mode_anim"] = True
params["mode_3D"] = True

recover_image(params)  # Main function to recover images
```

A way to have a quantitative idea of how accurate the reconstruction was is to take a look at the quality metrics (see output).
* K-Nearest Neighbors (KNN) evaluates the local structure by comparing neighbors in the original image with neighbors in the recovered image. It gives an overall idea on how well the local "fine grain" structure is preserved.
* Correlation Pairwise Distence (CPD) evaluates  the global structure by comparing distances in the original and recovered image. It gives an overall idea on how well the global "coarse grain" structure is preserved.

It is also interesting to know how much time it takes to recover an image and the amount of memory being used (time and space empirical complexity). Corresponding units are seconds for time and megabytes for memory.

Great! But how does the original image look like?

```python
params["embedding_mode"] = "original_image"  
params["manifold_learning_mode"] = "directly"
recover_image(params)
```

# Different approaches
It is possible to experiment with different algorithm combinations and explore the results both through visual observation or through our quantitative quality metrics.
For example, a quick glance at the implementation below reveals that a shortest path matrix reduced through PCA:
1. Has a strong global metric (value near one)
2. Has a "fish bowl" distortion, which slightly lowers the local quality metric compared to the previous example
3. Its memory consumption is high (as it stores a dense $5000 \times 5000$ matrix)

```python
params["embedding_mode"] = "shortest_paths" # Can be memory-consuming (don't try for system sizes > 10.000)
params["manifold_learning_mode"] = "PCA"
recover_image(params)
```

```python
params["embedding_mode"] = "spring_relaxation"  # Can be time-consuming
recover_image(params)
```

# Proximity graphs - Changing neighbor rules
The way DNA strands interact with one another may vary significantly from experiment to experiment. This causes the notion of "proximity" to change, that is, the very rules that define if two strands will interact and concatenate and become "neighbors". Therefore, there a robust approach that performs well under several scenarios is relevant to account for different experimental settings.

The default proximity graph that has been reconstructed so far is the "KNN graph". Let's try the "Delaunay graph". 
1. Shortest path matrix reduced through PCA
2. Structural embedding reduced through UMAP (default and suggested approach)

![Delaunay Graph](KNN_graph.png)  ![Delaunay Graph](Delaunay_graph.png)


```python
params["embedding_mode"] = "shortest_paths"
params["manifold_learning_mode"] = "PCA"
params["proximity_mode"] = "delaunay"
recover_image(params)
```

```python
params["embedding_mode"] = "node2vec" 
params["manifold_learning_mode"] = "UMAP"
params["proximity_mode"] = "delaunay"
recover_image(params)
```

# Use your own data!
As a user you can test your own data as input. The usage should be simple: introduce an "edge list" file under the folder 

`/Input_Documents/Edge_Lists`
    
with the format:

| $n_i$      | $n_j$ |
| ----------- | ----------- |
| 1      | 2       |
| 1      | 3       |
| 2      | 1       |
| 2      | 5       |
| ...      | ...       |
| N      | ...       |
 
Where each row represents a node-to-node proximity connection. 
To activate this mode the "proximity_mode" has to be set to "experimental". The resulting positions can be found under

`/Output_Documents/Embedded_Positions`

```python
default_parameters = Im_Rec_Arguments()
default_parameters.proximity_mode = "experimental"
default_parameters.title_edge_list_experimental = "edge_list_knn_1000.csv" # Make sure that your file exists in the folder /Input_Documents/Edge_Lists
params = default_parameters.__dict__
recover_image(params)
```
